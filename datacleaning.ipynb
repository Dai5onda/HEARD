```python
import pandas as pd

# Read the data from the CSV files
accel_L = pd.read_csv('data/accel_L.csv')
accel_R = pd.read_csv('data/accel_R.csv')
labelled_behaviors = pd.read_csv('data/labelled_behaviors.csv')
```

```python
labelled_behaviors['StartTime'] = labelled_behaviors['StartTime'].round(3)
accel_L['Time'] = accel_L['Time'].round(3)
accel_R['Time'] = accel_R['Time'].round(3)
# labelled_behaviors.head(5)
# accel_L.head(5)
# accel_R.head(5)
```

```python
labelled_behaviors = labelled_behaviors.sort_values('StartTime')
labelled_behaviors.head(42)
```

```python
labelled_behaviors['ActivityType'] = labelled_behaviors['ActivityType'].str.strip()
activity_types = labelled_behaviors['ActivityType'].value_counts()
num_activity_types = len(activity_types)

print("Number of unique ActivityTypes:", num_activity_types)
print("ActivityTypes:")
print(activity_types.index.tolist())
```

```python
labelled_behaviors.head(50)
labelled_behaviors = labelled_behaviors.reset_index(drop=True)
labelled_behaviors.head(50)
```

```python
df1 = pd.DataFrame(accel_L.head(5))
df2 = pd.DataFrame(accel_R.head(5))
df3 = pd.DataFrame(labelled_behaviors.head(5))
df1, df2, df3   
```

```python
import seaborn as sns
print(accel_R.isnull().sum())
print(accel_L.isnull().sum())
print(labelled_behaviors.isnull().sum())

import matplotlib.pyplot as plt

# Create a heatmap of missing values in accel_R dataframe
plt.figure(figsize=(10, 6))
sns.heatmap(accel_R.isnull(), cmap='viridis', cbar=False)
plt.title('Missing Values in accel_R')
plt.show()

# Create a heatmap of missing values in accel_L dataframe
plt.figure(figsize=(10, 6))
sns.heatmap(accel_L.isnull(), cmap='viridis', cbar=False)
plt.title('Missing Values in accel_L')
plt.show()

# Create a heatmap of missing values in labelled_behaviors dataframe
plt.figure(figsize=(10, 6))
sns.heatmap(labelled_behaviors.isnull(), cmap='viridis', cbar=False)
plt.title('Missing Values in labelled_behaviors')
plt.show()
```

```python
labelled_behaviors.to_csv('labelled_behaviors.csv', index=False)
accel_L.to_csv('accel_L.csv', index=False)
accel_R.to_csv('accel_R.csv', index=False)
```
